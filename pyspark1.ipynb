```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *

spark=SparkSession.builder.appName('UserBehavior').getOrCreate()
ubRdd = sc.textFile('UserBehavior.csv')
schema = StructType([
    StructField('user_id',IntegerType(),True),
    StructField('item_id',IntegerType(),True),
    StructField('cat_id',IntegerType(),True),
    StructField('be_type',StringType(),True),
    StructField('timestamp',IntegerType(),True),
])

sample_rdd = ubRdd.sample(False,0.01,456)
sample_rdd.count()
ubdf = spark.createDataFrame(sample_rdd)
ubdf.createOrReplaceTempView("UserBehaviorCsv")
#ubdf.count()

```
